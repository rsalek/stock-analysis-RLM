```python
import os
import sys
from pathlib import Path

repo_root = Path.cwd()
if (repo_root / 'phase0_rlm').exists():
    sys.path.insert(0, repo_root.as_posix())

from phase0_rlm import bm25_index, chunking, document_store, tools

os.environ['PHASE0_BASE_DRIVE_PATH'] = '/content/drive/My Drive/Stock_Analysis'
sample_ticker = 'DEMO'
input_dir = Path(os.environ['PHASE0_BASE_DRIVE_PATH']) / 'Input' / sample_ticker
document_store.ingest_drive_paths([input_dir])
chunking.chunk_documents()
bm25_index.build_bm25_index()
results = tools.search(f'{sample_ticker} overview', k=12, ticker=sample_ticker)
print(results[:3])
```
